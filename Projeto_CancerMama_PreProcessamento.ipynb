#Pre Processamento
Analisando o dataset e realizando modificações em caso de necessidade

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
```

```python
url = 'https://raw.githubusercontent.com/VictorHNascimento/Analise-Classificacao-de-Cancer-de-Mama/main/breast-cancer.csv'
!wget {url} -O breastcancer.csv
df = pd.read_csv('breastcancer.csv')
```

```python
df.head()
```

```python
#Analisando o tipo de dado da tabela
df.info()
```

```python
df.describe()
```

```python
#Verificando se temos dados nulos
df.isnull().sum()
```

```python
#Ajustando o Dataset
coluna = df.pop('diagnosis')
df['diagnosis'] = coluna
df['diagnosis'] = df['diagnosis'].map({'B': 0, 'M': 1})
df.head()
```

```python
df.to_csv('breast_cancer_ajustado.csv', index = False)
```

```python
#Verificando Contagem de Categorias e Total
contagem = df['diagnosis'].value_counts()
print(f'Valores de cada categoria:\n{contagem}')

print('Total de Dados:', len(df))
```

```python
#Verificando Outliers
import seaborn as sns

features = df.iloc[: , 1:-1]

# Exemplo de boxplot para identificar outliers em uma coluna
for i in features.columns:
  sns.boxplot(x=features[i])
  plt.show()
  print()
```

```python
#Verificando a Matriz de Correlação
correlation_matrix = features.corr()

# Configurar o tamanho da figura
plt.figure(figsize=(80, 64))

# Criar o heatmap da matriz de correlação
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Adicionar título ao gráfico
plt.title('Matriz de Correlação')

# Mostrar o gráfico
plt.show()
```

```python
#Verificando o Ganho de Informação
from sklearn.feature_selection import mutual_info_classif

np.random.seed(42)
data = df

# Lista para armazenar o ganho de informação de cada feature
ganho_informacao = []

# Selecione as features
for i in data.columns:
    if i != 'diagnosis':  # Exclua a coluna de target 'rating'
        X = data[[i]]  # Seleciona a feature
        y = data['diagnosis']  # Seleciona a label

        # Converta a feature 'age' para array NumPy (já que só tem uma)
        X_array = X.to_numpy()

        # Converta a label 'loan_status' para série Pandas usando 'pd.Series'
        y_series = pd.Series(y)

        # Calcule o Ganho de Informação usando mutual_info_classif
        ganho = mutual_info_classif(X_array, y_series)

        # Adicione o resultado à lista como uma tupla (nome da feature, ganho de informação)
        ganho_informacao.append((i, ganho[0]))

# Ordene a lista pelo ganho de informação em ordem decrescente
ganho_informacao = sorted(ganho_informacao, key=lambda x: x[1], reverse=True)

# Imprima o ranking do ganho de informação
print("Ranking do Ganho de Informação:")
for rank, (feature, info_gain) in enumerate(ganho_informacao, start=1):
    print(f"{rank}. {feature}: {info_gain:.2f}")
```
